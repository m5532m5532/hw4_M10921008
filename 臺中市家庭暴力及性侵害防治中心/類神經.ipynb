```python
X = df_view.drop('Y', axis=1)
y = df_view['Y']
```

```python
from tensorflow.keras.layers import Input, Dense, Activation,Dropout,BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping,ReduceLROnPlateau
from tensorflow.keras import optimizers
```

```python
trainx=X.values
trainy= to_categorical(y -1, 3)
```

```python
trainx.shape
```

```python
"""
å„ªåŒ–å™¨
"""
unit = 128
es = EarlyStopping(monitor='acc', patience=10, verbose=0, mode='max', baseline=None, restore_best_weights=False)
reduce_lr = ReduceLROnPlateau(monitor='acc', factor=0.1,mode='max',patience=5, min_lr=1e-07)
input_layer = Input(shape=(trainx.shape[1],))
output = Dense(unit, activation='sigmoid')(input_layer)
output = Dense(unit/2, activation='sigmoid')(output)
output = Dense(unit/4, activation='sigmoid')(output)
output = Dense(unit/8, activation='sigmoid')(output)
output = Dense(unit/16, activation='sigmoid')(output)
output = Dense(3, activation='softmax')(output)
model = Model(inputs=input_layer, outputs=output)
model.compile(loss="CategoricalCrossentropy" , optimizer=optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-07), metrics=["acc"])
```

```python
model.summary()
```

```python
history = model.fit(trainx, trainy, batch_size=16, epochs=100, verbose=1,callbacks=([reduce_lr],[es]))
```

```python
df_test = pd.read_csv("test.csv")
df_test
```

```python
df_Test = df_test.drop(['ID'],axis=1)
df_Test
```

```python
answer_X = model.predict(df_Test.values)
```

```python
answer_result = pd.DataFrame()
answer_result = pd.DataFrame(answer_X,columns=['C1','C2','C3'])
answer_result['ID'] = df_test['ID']
answer_result.to_csv("results.csv",index=False)
answer_result
```
