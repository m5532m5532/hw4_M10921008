```python
#正規化(0~1)
def normalize_pandas(data):
    from sklearn.preprocessing import MinMaxScaler
    scaler = MinMaxScaler()
    scaler.fit(data)
    data[data.columns] = scaler.fit_transform(data[data.columns].to_numpy()) 
    return data
```

```python
train_infoNor = normalize_pandas(train_info)
train_infoNor
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler_ans = MinMaxScaler()
train_ansNor = scaler_ans.fit(train_ans.values.reshape(-1,1))
train_ansNor1 = scaler_ans.fit_transform(train_ans.values.reshape(-1,1))
train_ansNor2=np.ravel(train_ansNor1)
```

```python
predict_2Nor = normalize_pandas(predict_2)
predict_2Nor
```

## 隨機森林

```python
from sklearn.ensemble import RandomForestRegressor
# 建立 random forest 模型
rf = RandomForestRegressor(n_estimators = 1000,
                           n_jobs = -1,
                           oob_score = True,
                           bootstrap = True,
                           random_state = 41)
rf.fit(train_infoNor, train_ansNor2)
# 預測
train_infoNor_pred = rf.predict(train_infoNor)
```

```python
print('R^2 Training Score: {:.2f} \nOOB Score: {:.2f} \nR^2 Validation Score: {:.2f}'.format(rf.score(train_infoNor, train_ansNor2),rf.oob_score_,rf.score(train_infoNor, train_ansNor2)))
```

```python
# 特徵的重要度
rf.feature_importances_
df_feature=train_infoNor.columns
df_important = pd.DataFrame(index=df_feature)
df_important['values'] = rf.feature_importances_
df_important.plot.bar()
```

```python
#訓練
from sklearn.metrics import mean_squared_error
print('RMSE为：',np.sqrt(mean_squared_error(train_ansNor2,train_infoNor_pred)))
import scipy
import sklearn
from sklearn.model_selection import train_test_split
# from sklearn.metrics import mean_absolute_percentage_error
print('MAPE为：',mean_squared_error(train_ansNor2,train_infoNor_pred))
```

## 類神經

```python
from tensorflow.keras.layers import Input, Dense, Activation,Dropout,BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping,ReduceLROnPlateau
from tensorflow.keras import optimizers
#網路層
input_layer = Input(shape=(train_infoNor.shape[1],))
dense_layer_1 = Dense(32, activation='sigmoid')(input_layer)
dense_layer_2 = Dense(16, activation='sigmoid')(dense_layer_1)
dense_layer_3 = Dense(8, activation='sigmoid')(dense_layer_2)
output = Dense(1)(dense_layer_3)
#優化器
es = EarlyStopping(monitor='mse', patience=10, verbose=0, mode='max', baseline=None, restore_best_weights=False)
reduce_lr = ReduceLROnPlateau(monitor='mse', factor=0.1,mode='max',patience=5, min_lr=1e-07)
model = Model(inputs=input_layer, outputs=output)
model.compile(loss="mean_squared_error" , optimizer=optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-07), metrics=["mse"])
print(model.summary())
history = model.fit(train_infoNor, train_ansNor2, batch_size=2, epochs=30, verbose=1,callbacks=([reduce_lr],[es]))
```

```python
# 預測
df_test_y_NNpredicted = model.predict(train_infoNor)
print('RMSE为：',np.sqrt(mean_squared_error(df_test_y_NNpredicted, train_ansNor2)))
print('MAPE为：',mean_squared_error(df_test_y_NNpredicted, train_ansNor2))
```
